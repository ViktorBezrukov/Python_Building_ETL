## Data Cleansing and Transformation in Python

```python
import pandas as pd
pd.set_option('display.max_columns', None)
df_crashes = pd.read_csv("data/traffic_crashes.csv")
df_crashes.head()
```

```python
df_crashes.info()
```

```python
df_crashes.notnull().sum()
```

```python
df_crashes.isnull().sum()
```

```python
df_crashes.dropna(axis='columns', how='all', inplace=True)
```

```python
df_crashes.info()
```

```python
# Dropping rows with too many missing values
df_crashes = df_crashes.dropna(axis='index', thresh=2,
inplace=False)
```

```python
# Working with missing data

# This column has only two values.
df_crashes['report_type'].unique
# ['ON SCENE', 'NOT ON SCENE (DESK REPORT)', nan]
# Length: 3, dtype: str
# Let's fill the missing value with 'ON SCENE' as below
df_crashes = df_crashes.fillna(value={'report_type': 'ON SCENE'})
```

```python
# Merging datasets and data aggregation
df_vehicles= pd.read_csv("data/traffic_crash_vehicle.csv")
df_vehicles.head()
```

```python
df = df_crashes.merge(df_vehicles, how = 'left', on='crash_record_id',suffixes=('_left', '_right'))
```

```python
df.head()
```

```python
df.info()
```

```python
df_agg = df.groupby('vehicle_type').agg({'crash_record_id':'count'}).reset_index()
```

```python
df_agg
```

```python
number_of_passenger_cars_involved = df_agg[df_agg['vehicle_type'] ==
'PASSENGER']['crash_record_id'].array[0]
```

```python
number_of_passenger_cars_involved
```

```python
vehicle_mapping = {'vehicle_type' : 'vehicletypes'}
df_agg = df_agg.rename(columns=vehicle_mapping)
```
